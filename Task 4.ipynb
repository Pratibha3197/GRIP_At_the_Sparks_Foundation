###### 1)Imprting the libraries

```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets
```

###### 2)Importing Datasets

```python
iris = datasets.load_iris()
```

```python
X=pd.DataFrame(iris.data,columns = iris.feature_names)
X.head()
```

```python
Y=pd.DataFrame(iris.target)
Y.head()
```

###### 3)Training the model

```python
from sklearn import tree
classifier = tree.DecisionTreeClassifier()
classifier.fit(X,Y)
```

###### 4)Decision Tree Visualisation

```python
fig,axes = plt.subplots(nrows=1,ncols=1,figsize = (10,10),dpi = 500)
tree.plot_tree(classifier,feature_names=iris.feature_names,class_names=iris.target_names,filled = True)
fig.savefig('DecisionTree.png')
```

###### 5)Accuracy of the model

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2,random_state = 0)
```

```python
from sklearn.metrics import accuracy_score
from sklearn import tree
clf = tree.DecisionTreeClassifier()
clf.fit(X_train,Y_train)
Y_pred = clf.predict(X_test)
accuracy_score(Y_test,Y_pred)
```
